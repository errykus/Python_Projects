<b> Q3. The Job Role Columns holds the job information for each individual in this census snapshot. Using this column, create a Bar Chart that shows the count of 'Unique' Jobs per Job Group in the "Job-Role" Column in ascending order, as per the provided image below </b>


<b> Put your code below </b>

```python

job_counts = df['Job-Role'].value_counts().sort_values(ascending=True)

plt.bar(job_counts.index, job_counts.values)
plt.xlabel('Job Group')
plt.ylabel('Count')
plt.title('Count of Unique Jobs per Job Group')
plt.xticks(rotation=90)
plt.show()
```

<b> Q4. Please create two bar plots as per below that show:
    
    1) The number of individuals who have a High School Graduate Diploma AND earn <=50K in the United States
    2) The number of individuals who have a High School Graduate Diploma AND earn >50K in the United States 

Please note you will be looking specifically at the *Job Role* column

<b> Put Your Code Below </b>

```python
df['Earnings'].unique()
```

```python
filtered_df_1 = df[(df['Degree Status'] ==' HS-grad') & (df['Earnings'] == ' <=50K')]
filtered_df_2 = df[(df['Degree Status'] ==' HS-grad') & (df['Earnings'] == ' >50K')]

job_counts1 = filtered_df_1['Job-Role'].value_counts().sort_values(ascending=False)
job_counts2 = filtered_df_2['Job-Role'].value_counts().sort_values(ascending=False)


# Create two subplots for each chart
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Chart 1: Individuals earning <=50K
ax1.bar(job_counts1.index, job_counts1.values)
ax1.tick_params(axis='x', rotation=90)  # Rotate x-axis labels

# Chart 2: Individuals earning >50K
ax2.bar(job_counts2.index, job_counts2.values)


# Adjust spacing between subplots
plt.tight_layout()
ax2.tick_params(axis='x', rotation=90)  # Rotate x-axis labels

# Display the charts
plt.show()
```

<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
df_high_earnings = df[df['Earnings'] == ' >50K']

proportions = df_high_earnings['Job-Role'].value_counts(normalize=True)

# Get the Job Role with the highest proportion
highest_proportion_job_role = proportions.idxmax()

print(f"The Job Role with the highest proportion of individuals earning >50K is: {highest_proportion_job_role}")
```
